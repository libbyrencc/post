### Trying to visualize the body mass and the area of Culmen

```python
import pandas as pd
url = "https://raw.githubusercontent.com/PhilChodrow/PIC16B/master/datasets/palmer_penguins.csv"
penguins = pd.read_csv(url)
```

Read 5 lines

```python
penguins.head(5)
```

Notice that there are penguins with NaN, we need to drop that.

```python
penguins=penguins.dropna(subset=['Culmen Length (mm)','Culmen Depth (mm)','Body Mass (g)'])
penguins.head(5)
```

calculate the area of penguins' culmen

```python
penguins['Culmen Area (mm^2)']=penguins['Culmen Length (mm)']*penguins['Culmen Depth (mm)']/2
```

```python
penguins.head(5)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.scatterplot(data = penguins, 
                x = "Culmen Area (mm^2)", 
                y = "Body Mass (g)", 
                hue = "Species",
                palette = "vlag")
plt.legend(bbox_to_anchor=(1.05, 1),loc=2)
sns.despine()
```
